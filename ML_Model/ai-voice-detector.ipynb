```python
import librosa

audio_path = "//kaggle/input/datasets/awsaf49/asvpoof-2019-dataset/LA/LA/ASVspoof2019_LA_train/flac/LA_T_1000137.flac"

audio, sr = librosa.load(audio_path, sr=None)
print(sr)
print(audio.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
plt.plot(audio)
plt.title("Audio Waveform")
plt.show()
```

```python
import numpy as np
mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
mfcc = np.mean(mfcc.T, axis=0)
print(mfcc)
print(mfcc.shape)
```
